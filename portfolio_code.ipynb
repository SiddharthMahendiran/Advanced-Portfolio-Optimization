```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta

# Get the list of Nifty 50 stocks from NSE's Wikipedia page or an existing list
nifty50_stocks = [
    "RELIANCE.NS", "HDFCBANK.NS", "INFY.NS", "ICICIBANK.NS", "TCS.NS", "KOTAKBANK.NS", "ITC.NS", "LT.NS", "SBIN.NS",
    "AXISBANK.NS", "BHARTIARTL.NS", "HINDUNILVR.NS", "ASIANPAINT.NS", "BAJFINANCE.NS",
    "MARUTI.NS", "HCLTECH.NS", "TITAN.NS", "SUNPHARMA.NS", "WIPRO.NS",
    "ULTRACEMCO.NS", "ONGC.NS", "NTPC.NS", "POWERGRID.NS", "TATASTEEL.NS",
    "ADANIGREEN.NS", "HINDALCO.NS", "JSWSTEEL.NS", "COALINDIA.NS", "BPCL.NS",
    "NESTLEIND.NS", "M&M.NS", "INDUSINDBK.NS", "DIVISLAB.NS", "ADANIPORTS.NS",
    "GRASIM.NS", "EICHERMOT.NS", "HEROMOTOCO.NS", "BAJAJ-AUTO.NS", "TECHM.NS",
    "DRREDDY.NS", "APOLLOHOSP.NS", "CIPLA.NS", "BRITANNIA.NS", "SHREECEM.NS",
    "TATAMOTORS.NS", "ADANIENT.NS", "UPL.NS", "SBILIFE.NS", "BAJAJFINSV.NS"
]

# Calculate the date range for the previous two years
end_date = datetime.now()
start_date = end_date - timedelta(days=2*365)

# Create a dictionary to hold the data
closing_prices = {}

# Loop through each stock symbol
for stock in nifty50_stocks:
    # Download historical data for the stock
    data = yf.download(stock, start=start_date, end=end_date)
    
    # Store the closing prices in the dictionary
    closing_prices[stock] = data['Close']

# Convert the dictionary to a DataFrame
closing_prices_df = pd.DataFrame(closing_prices)

# Display the DataFrame
print(closing_prices_df)
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
from tensorflow.keras.optimizers import Adam

# Assuming closing_prices_df is already loaded with the past 2 years of closing prices
# Columns: Stock names, Rows: Date index with daily closing prices

# Drop any rows with NaN values to ensure clean data
closing_prices_df.dropna(inplace=True)

# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(closing_prices_df)

# Set the look-back period (e.g., 60 days)
look_back_period = 60

# Function to create sequences for LSTM
def create_sequences(data, look_back):
    X, y = [], []
    for i in range(len(data) - look_back):
        X.append(data[i:i + look_back])
        y.append(data[i + look_back])
    return np.array(X), np.array(y)

# Create sequences
X, y = create_sequences(scaled_data, look_back_period)

# Reshape input data to fit LSTM [samples, time steps, features]
X = X.reshape(X.shape[0], X.shape[1], X.shape[2])

# Split the data into training and testing sets (80% train, 20% test)
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# Build the LSTM model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(look_back_period, X_train.shape[2])))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(25))
model.add(Dense(X_train.shape[2]))  # Output one value per stock

# Compile the model
model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, batch_size=32, epochs=20, validation_data=(X_test, y_test))

# Predict future prices for the next 60 days (3 months)
predicted_future_prices = []

# Start with the last sequence from the test data
last_sequence = X_test[-1]

for _ in range(60):  # Predict for 60 trading days (3 months)
    next_day_prediction = model.predict(last_sequence.reshape(1, look_back_period, X_train.shape[2]))
    predicted_future_prices.append(next_day_prediction.flatten())
    
    # Update the sequence: remove the oldest price and add the predicted price
    last_sequence = np.append(last_sequence[1:], next_day_prediction, axis=0)

# Inverse transform the predictions to get the actual prices
predicted_future_prices = np.array(predicted_future_prices)
predicted_future_prices = scaler.inverse_transform(predicted_future_prices)

# Create a DataFrame for the predicted future prices
future_dates = pd.date_range(start=closing_prices_df.index[-1], periods=61, freq='B')[1:]
predicted_future_df = pd.DataFrame(predicted_future_prices, index=future_dates, columns=closing_prices_df.columns)

print("Predicted stock prices for the next 3 months:")
print(predicted_future_df)
```

```python
# Get the last known prices (current prices)
current_prices = closing_prices_df.iloc[-1].values

# Get the predicted prices for 60 days into the future
forecasted_prices_60_days = predicted_future_df.iloc[-1].values

# Calculate the return for each stock
returns = (forecasted_prices_60_days - current_prices) / current_prices * 100

# Create a DataFrame for the returns
returns_df = pd.DataFrame({
    'Stock': closing_prices_df.columns,
    'Current Price': current_prices,
    'Forecasted Price (60 days)': forecasted_prices_60_days,
    '60-Day Return (%)': returns
})

print("Returns for each stock over the next 60 days:")
returns_df
```

```python
# Recalculate daily returns and annualize them if needed
daily_returns = closing_prices_df.pct_change().dropna()
annualized_volatility = daily_returns.std() * np.sqrt(252)  # Assuming 252 trading days in a year

# Recalculate the covariance matrix and ensure it's positive semi-definite
historical_covariance = daily_returns.cov() * 252

# Normalize future returns to annualized returns if needed
annualized_returns = returns / 60 * 252  # Converting 60-day returns to annualized

# Recalculate the Sharpe Ratio
num_portfolios = 10000
results = np.zeros((3, num_portfolios))
weights_record = []

for i in range(num_portfolios):
    positive_returns = annualized_returns > 0
    weights = np.random.random(len(closing_prices_df.columns))
    weights[~positive_returns] = 0
    weights /= np.sum(weights)
    
    weights_record.append(weights)
    
    portfolio_return = np.sum(annualized_returns * weights)
    portfolio_volatility = (np.sqrt(np.dot(weights.T, np.dot(historical_covariance, weights))))*100
    
    sharpe_ratio = portfolio_return / portfolio_volatility
    
    results[0, i] = portfolio_return
    results[1, i] = portfolio_volatility
    results[2, i] = sharpe_ratio

max_sharpe_idx = np.argmax(results[2])
max_sharpe_allocation = weights_record[max_sharpe_idx]

print("Portfolio with maximum Sharpe Ratio:")
print("Return:", results[0, max_sharpe_idx])
print("Volatility:", results[1, max_sharpe_idx])
print("Sharpe Ratio:", results[2, max_sharpe_idx])
```
